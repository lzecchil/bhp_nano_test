```python
import ROOT
from ROOT import RDataFrame
from ROOT import TChain, TSelector, TTree, TH1F, TCanvas, TFile, TEfficiency, TLegend
from ROOT import TLorentzVector
from ROOT import TVector3
from array import array
import numpy as np
ROOT.gROOT.SetBatch(ROOT.kTRUE)
file ="../bph_nano_test/bph_nano_test_mc_TracksMuons.root"
f=TFile(file)
tree=f.Get("Events")
#tree.Print()
N = tree.GetEntries()
print("total number of events=",N)
```

```python
hNumK=TH1F("hNumK","K ricostruiti per pt",100,0,200)
hNumpi=TH1F("hNumpi","pi ricostruiti per pt",100,0,200)
hNum=TH1F("hNum","K e pi ricostruiti per pt",100,0,200)
hDen=TH1F("hDen","K e pi generati per pt",100,0,200)
hDenK=TH1F("hDenK","K generati per pt",100,0,200)
hDenpi=TH1F("hDenpi","pi generati per pt",100,0,200)
hdR=TH1F("hdR","distribuzione dR",1000,0,1)            
hdRK=TH1F("hdRK","distribuzione dR dei K",1000,0,1)
hdRpi=TH1F("hdRpi","distribuzione dR dei pi",1000,0,1)
        
```

```python

hNumK.Reset()
hNumpi.Reset()
hNum.Reset()
hDen.Reset()
hDenK.Reset()
hDenpi.Reset()
hdR.Reset()
hdRK.Reset()
hdRpi.Reset()
pdg = ROOT.TDatabasePDG.Instance()
for i in range(N):
    tree.GetEntry(i)
    for id,midx,pt,eta,phi,st,mass in zip(tree.GenPart_pdgId,tree.GenPart_genPartIdxMother,tree.GenPart_pt,tree.GenPart_eta,tree.GenPart_phi,tree.GenPart_status,tree.GenPart_mass):
        if abs(id) in (211,321) and abs(tree.GenPart_pdgId[midx]) in (333,310) and abs(tree.GenPart_pdgId[tree.GenPart_genPartIdxMother[midx]])==511 and st==1:
            GenP=TLorentzVector()
            hDen.Fill(pt)
            if abs(id)==211:
                hDenpi.Fill(pt)
                GenP.SetPtEtaPhiM(pt,eta,phi,pdg.GetParticle(id).Mass())
            else:
                hDenK.Fill(pt)
                GenP.SetPtEtaPhiM(pt,eta,phi,mass)
            dR_min=1000
            for t_pt,t_eta,t_phi,t_mass in zip(tree.Track_pt,tree.Track_eta,tree.Track_phi,tree.Track_mass):
                Track=TLorentzVector()
                Track.SetPtEtaPhiM(t_pt,t_eta,t_phi,t_mass)
                dR=GenP.DeltaR(Track)
                if dR<dR_min:
                    dR_min=dR
            hdR.Fill(dR_min)
            if abs(id)==211:
                hdRpi.Fill(dR_min)
            else:
                hdRK.Fill(dR_min)           
            if abs(id)==211 and dR_min<0.1:
                hNumpi.Fill(pt)
                hNum.Fill(pt)
            elif abs(id)==321 and dR_min<0.02:
                hNumK.Fill(pt)
                hNum.Fill(pt)

            
```

```python
h=TH1F("h","Efficiency(K) vs pt",100,0,200)
h.Reset()
c1=TCanvas('c1','Efficiency K vs Pt',200,10,700,500)
c1.cd()
EffK_pt=TEfficiency(hNumK,hDenK)
h.GetXaxis().SetTitle("Gen K p_{t} [Gev/c]")
h.GetYaxis().SetTitle("Efficiency")
ROOT.gStyle.SetOptStat("e")
h.SetStats(0)
h.Draw()
EffK_pt.Draw("same")
c1.Draw()
```

```python
h2=TH1F("h2","Efficiency(pi) vs pt",100,0,200)
h2.Reset()
c2=TCanvas('c2','Efficiency pi vs Pt',200,10,700,500)
c2.cd()
Effpi_pt=TEfficiency(hNumpi,hDenpi)
h2.GetXaxis().SetTitle("Gen pi p_{t} [Gev/c]")
h2.GetYaxis().SetTitle("Efficiency")
h2.SetStats(0)
h2.Draw()
Effpi_pt.Draw("same")
c2.Draw()
```

```python
h3=TH1F("h3","Efficiency vs pt",100,0,200)
h3.Reset()
c3=TCanvas('c3','Efficiency_Pt',200,10,700,500)
c3.cd()
Eff_pt=TEfficiency(hNum,hDen)
h3.GetXaxis().SetTitle("Gen pi/K p_{t} [Gev/c]")
h3.GetYaxis().SetTitle("Efficiency")
h3.SetStats(0)
h3.Draw()
Eff_pt.Draw("same")
c3.Draw()
    
```

```python
c5=TCanvas("c5","K",200,10,700,500)
c5.cd()
hNumK.SetLineColor(2)
hDenK.SetLineColor(4)
hNumK.Draw()
hDenK.Draw("same")
hNumK.SetTitle("Gen K compared to matched gen K vs pt")
hNumK.GetXaxis().SetTitle("Gen K p_{t} [Gev/c]")
hNumK.GetYaxis().SetTitle("number of Kaons")
legend=TLegend(0.6,0.8,0.85,0.6)
legend.AddEntry(hNumK,"Gen K")
legend.AddEntry(hDenK,"matched gen K")
ROOT.gStyle.SetLegendTextSize(0.05)
legend.Draw()
c5.Draw()
```

```python
c6=TCanvas("c6","pi",200,10,700,500)
c6.cd()
hNumpi.SetLineColor(2)
hDenpi.SetLineColor(4)
hNumpi.Draw()
hDenpi.Draw("same")
hNumpi.SetTitle("Gen pi compared to matched gen pi vs pt")
hNumpi.GetXaxis().SetTitle("Gen pi p_{t} [Gev/c]")
hNumpi.GetYaxis().SetTitle("number of pions")
legend=TLegend(0.6,0.8,0.85,0.6)
legend.AddEntry(hNumpi,"Gen pi")
legend.AddEntry(hDenpi,"matched gen pi")
ROOT.gStyle.SetLegendTextSize(0.05)
legend.Draw()
c6.Draw()
```

```python
c3=TCanvas('c3','Efficiency_Pt',200,10,700,500)
c3.cd()
Eff_pt=TEfficiency(hNum,hDen)
Eff_pt.SetTitle("Efficiency vs Pt")
Eff_pt.Draw()
ROOT.gPad.Update()
graph=Eff_pt.GetPaintedGraph()
if graph:
    graph.GetXaxis().SetTitle("Gen pi\K p_{t} [Gev/c]")
    graph.GetYaxis().SetTitle("Efficiency")
c3.Draw()
```
