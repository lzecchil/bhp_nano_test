```python
import ROOT
from ROOT import RDataFrame
from ROOT import TChain, TSelector, TTree, TH1F, TCanvas, TFile, TEfficiency, TLegend
from ROOT import TLorentzVector
from ROOT import TVector3
from array import array
import numpy as np
ROOT.gROOT.SetBatch(ROOT.kTRUE)
file ="../bph_nano_test/bph_nano_test_mc_TracksMuons.root"
f=TFile(file)
tree=f.Get("Events")
#tree.Print()
N = tree.GetEntries()
print("total number of events=",N)
```

```python
hNumK=TH1F("hNumK","reco K vs pt",100,0,50)
hNumKp=TH1F("hNumKp","reco K+ vs pt",100,0,50)
hNumKm=TH1F("hNumKm","reco K- vs pt",100,0,50)
hNumpi=TH1F("hNumpi","reco pi vs pt",100,0,50)
hNumpip=TH1F("hNumpip","reco pi+ vs pt",100,0,50)
hNumpim=TH1F("hNumpim","reco pi- vs pt",100,0,50)
hNum=TH1F("hNum","reco K/pi vs pt",100,0,50)
hDen=TH1F("hDen","gen K/pi vs pt",100,0,50)
hDenK=TH1F("hDenK","gen K vs pt",100,0,50)
hDenKp=TH1F("hDenKp","gen K+ vs pt",100,0,50)
hDenKm=TH1F("hDenKm","gen K- vs pt",100,0,50)
hDenpi=TH1F("hDenpi","gen pi vs pt",100,0,50)
hDenpip=TH1F("hDenpip","gen pi+ vs pt",100,0,50)
hDenpim=TH1F("hDenpim","gen pi- vs pt",100,0,50)
hdR=TH1F("hdR","dR distribution",1000,0,1)            
hdRK=TH1F("hdRK","dR distribution (K)",1000,0,1)
hdRpi=TH1F("hdRpi","dR distribution (pi)",1000,0,1)

        
```

```python
def matching(th,charge,GenP,pt,tree,h1,h2,h3,h4,h5):
    Result=False
    dR_min=1000
    for ch,t_pt,t_eta,t_phi,t_mass in zip(tree.Track_charge,tree.Track_pt,tree.Track_eta,tree.Track_phi,tree.Track_mass):
        if ch==charge:
            Track=TLorentzVector()
            Track.SetPtEtaPhiM(t_pt,t_eta,t_phi,t_mass)
            dR=GenP.DeltaR(Track)
            if dR<dR_min:
                dR_min=dR
    h4.Fill(dR_min)
    h5.Fill(dR_min)
    if dR_min<th:
        Result=True
        h1.Fill(pt)
        h2.Fill(pt)
        h3.Fill(pt)
```

```python

hNumK.Reset()
hNumKp.Reset()
hNumKm.Reset()
hNumpi.Reset()
hNumpip.Reset()
hNumpim.Reset()
hNum.Reset()
hDen.Reset()
hDenK.Reset()
hDenKp.Reset()
hDenKm.Reset()
hDenpi.Reset()
hDenpip.Reset()
hDenpim.Reset()
hdR.Reset()
hdRK.Reset()
hdRpi.Reset()
pdg = ROOT.TDatabasePDG.Instance()
for i in range(N):
    tree.GetEntry(i)
    for id,midx,pt,eta,phi,st,mass in zip(tree.GenPart_pdgId,tree.GenPart_genPartIdxMother,tree.GenPart_pt,tree.GenPart_eta,tree.GenPart_phi,tree.GenPart_status,tree.GenPart_mass):
        gmidx=tree.GenPart_genPartIdxMother[midx]
        if ((abs(id)==211 and abs(tree.GenPart_pdgId[midx])==310) or (abs(id)==321 and abs(tree.GenPart_pdgId[midx])==333 )) and abs(tree.GenPart_pdgId[gmidx])==511 and st==1:
            GenP=TLorentzVector()
            hDen.Fill(pt)
            if abs(id)==211:
                GenP.SetPtEtaPhiM(pt,eta,phi,pdg.GetParticle(id).Mass())
                hDenpi.Fill(pt)
                if id<0:
                    hDenpim.Fill(pt)
                    matching(0.1,-1,GenP,pt,tree,hNum,hNumpi,hNumpim,hdR,hdRpi)
                        
                else:
                    hDenpip.Fill(pt)
                    matching(0.1,1,GenP,pt,tree,hNum,hNumpi,hNumpip,hdR,hdRpi)
            else:
                hDenK.Fill(pt)
                GenP.SetPtEtaPhiM(pt,eta,phi,mass)
                if id<0:
                    hDenKm.Fill(pt)
                    matching(0.02,-1,GenP,pt,tree,hNum,hNumK,hNumKm,hdR,hdRK)
                else:
                    hDenKp.Fill(pt)
                    matching(0.02,1,GenP,pt,tree,hNum,hNumK,hNumKp,hdR,hdRK)
            
            
```

```python
h=TH1F("h","Efficiency(K) vs pt",100,0,50)
h.Reset()
c1=TCanvas('c1','Efficiency K vs Pt',200,10,700,500)
c1.cd()
EffK_pt=TEfficiency(hNumK,hDenK)
h.GetXaxis().SetTitle("Gen K p_{t} [Gev/c]")
h.GetYaxis().SetTitle("Efficiency")
ROOT.gStyle.SetOptStat("e")
h.SetStats(0)
h.Draw()
EffK_pt.Draw("same")
c1.Draw()
```
